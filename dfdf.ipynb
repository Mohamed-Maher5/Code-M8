```python
# llm/classifier.py
# Message classifier — determines if a user message is a task or chat
from core.config import GROQ_API_KEY, GROQ_MODEL
from utils.logger import logger
from groq import Groq

_CLASSIFIER_SYSTEM = _CLASSIFIER_SYSTEM = """You are a strict message classifier.

Your job is to decide if the user wants to DO something (task) or just TALK (chat).

task — the user wants action taken on code or files:
       write, fix, explain, read, edit, create, refactor, debug,
       implement, search, find, list, add, remove, update, run, etc.
       Also includes questions like "how do I..." or "can you show me..."
       when they expect code or a file operation as output.

chat — the user is just talking:
       greetings (hi, hello, hey), thanks, compliments, small talk,
       questions about you (who are you, what can you do),
       or any message that does NOT expect code or file output.

Examples:
  "hey what's up"              → chat
  "thanks!"                    → chat
  "what are you?"              → chat
  "fix the bug in main.py"     → task
  "explain this function"      → task
  "how do I read a file?"      → task
  "write me a sorting function"→ task

Reply with a single word only: task or chat. No punctuation. No explanation."""

def classify_message(user_message: str) -> str:
        client = Groq(api_key=GROQ_API_KEY)
        response = client.chat.completions.create(
            model=GROQ_MODEL,
            messages=[
                {"role": "system", "content": _CLASSIFIER_SYSTEM},
                {"role": "user", "content": user_message},
            ],
            max_tokens=1024,
            temperature=0.0,

            include_reasoning=False,
        )
        raw = response.choices[0].message.content
        print(raw)


        # strip thinking tokens if present
        
classify_message("hi")
```
